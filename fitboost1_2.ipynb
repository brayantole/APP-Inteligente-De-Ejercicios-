```python
#!/usr/bin/env python3
# FitBoost Pro CLI Colombia 🇨🇴
# Versión POO con clases y objetos

class Usuario:
    def __init__(self):
        self.nombre = ""
        self.sexo = None
        self.edad = 0
        self.peso_kg = 0.0
        self.altura_m = 0.0
        self.nivel_actividad = ""
        self.multiplicador_actividad = 1.0
        self.dias_entrenamiento = 0
        self.duracion_entrenamiento = 0
        self.objetivo = ""
        self.usar_suplementos = False
        self.presupuesto_suplementos = 0
        self.recordatorio_agua = False
        self.intervalo_agua = 0
        self.recordatorio_ejercicio = False
        self.intervalo_ejercicio = 0
        self.rutina_sugerida = ""

    def calcular_imc(self):
        return self.peso_kg / (self.altura_m * self.altura_m)

    def clasificar_imc(self):
        imc = self.calcular_imc()
        if imc < 18.5:
            return "Bajo peso"
        elif imc < 25:
            return "Normal"
        elif imc < 30:
            return "Sobrepeso"
        else:
            return "Obesidad"

    def calcular_calorias(self):
        altura_cm = self.altura_m * 100
        if self.sexo == 1:  # Hombre
            bmr = 10 * self.peso_kg + 6.25 * altura_cm - 5 * self.edad + 5
        else:  # Mujer
            bmr = 10 * self.peso_kg + 6.25 * altura_cm - 5 * self.edad - 161

        tdee = bmr * self.multiplicador_actividad

        if self.objetivo == "Bajar de peso":
            meta = max(1200, tdee - 500)
        elif self.objetivo == "Ganar masa muscular":
            meta = tdee + 300
        else:  # Definir
            meta = max(1200, tdee - 300)

        return tdee, meta


class Rutina:
    def __init__(self):
        self.rutinas = {
            "Full Body": {
                "Casa": """Rutina 1 – Full Body (Casa)
Objetivo: Hipertrofia general.
1) Flexiones de pecho — 4x10-15 (60s) — Peso corporal
2) Sentadillas con peso corporal — 4x15-20 (60s) — Peso corporal/backpack
3) Fondos entre sillas — 3x10-12 (60s) — Peso corporal
4) Remo invertido en mesa — 3x12-15 (60s) — Peso corporal
5) Zancadas alternas — 3x12 c/pierna (60s)
6) Plancha frontal — 3x30-45s
7) Elevaciones laterales con botellas — 3x12-15 (45s)
8) Crunch — 3x20""",
                "Gimnasio": """Rutina 1 – Full Body (Gimnasio)
Objetivo: Hipertrofia total con compuestos.
1) Press de banca con barra — 4x8-10 (90s)
2) Sentadilla con barra — 4x8-10 (90s)
3) Peso muerto convencional — 3x6-8 (120s)
4) Dominadas o jalón al pecho — 3x10-12 (90s)
5) Press militar con mancuernas — 3x10-12 (60s)
6) Curl de bíceps con barra — 3x12 (60s)
7) Extensión tríceps en polea — 3x12-15 (60s)
8) Elevaciones de talones — 3x15-20 (45s)
9) Crunch en banco — 3x20"""
            },
            "Upper/Lower": {
                "Casa": """Rutina 2 – Upper/Lower (Casa)
Día 1 — Upper:
1) Flexiones — 4x12-15
2) Fondos entre sillas — 3x10-12
3) Remo con mochila — 4x12
4) Elevaciones laterales — 3x15
5) Curl bíceps con botellas — 3x12-15
6) Plancha — 3x30-45s

Día 2 — Lower:
1) Sentadillas con mochila — 4x15-20
2) Zancadas — 4x12 c/pierna
3) Peso muerto rumano con mochila — 3x12-15
4) Puente de glúteos — 3x15-20
5) Elevaciones de talones — 3x20""",
                "Gimnasio": """Rutina 2 – Upper/Lower (Gimnasio)
Día 1 — Upper:
1) Press de banca — 4x8-10
2) Press militar — 3x10-12
3) Dominadas — 4x10-12
4) Remo con barra — 3x10-12
5) Curl bíceps — 3x12
6) Extensión tríceps — 3x12-15

Día 2 — Lower:
1) Sentadilla con barra — 4x8-10
2) Peso muerto rumano — 4x8-10
3) Prensa de piernas — 3x12
4) Extensión de piernas — 3x15
5) Curl femoral — 3x15
6) Elevación de talones — 3x20"""
            },
            "Arnold Split": {
                "Casa": """Rutina 3 – Arnold Split (Casa)
Día 1 — Pecho & Espalda:
1) Flexiones — 4x15
2) Flexiones pies elevados — 3x12
3) Remo con mochila — 4x12
4) Superman — 3x20s
5) Fondos — 3x12

Día 2 — Hombros & Brazos:
1) Elevaciones laterales — 4x15
2) Flexiones diamante — 3x12
3) Curl bíceps con botellas — 3x15
4) Elevaciones frontales — 3x15
5) Plancha con toques — 3x20 toques

Día 3 — Piernas & Abdomen:
1) Sentadillas con mochila — 4x20
2) Zancadas — 4x12 c/pierna
3) RDL con mochila — 3x15
4) Puente de glúteos — 3x20
5) Crunch bicicleta — 3x20
6) Plancha frontal — 3x45s""",
                "Gimnasio": """Rutina 3 – Arnold Split (Gimnasio)
Día 1 — Pecho & Espalda:
1) Press de banca — 4x8-10
2) Press inclinado mancuernas — 4x10-12
3) Aperturas — 3x12
4) Dominadas — 4x10-12
5) Remo con barra — 4x10

Día 2 — Hombros & Brazos:
1) Press militar — 4x8-10
2) Elevaciones laterales — 4x12-15
3) Pájaros — 3x12-15
4) Curl barra — 4x10-12
5) Extensión tríceps — 4x10-12

Día 3 — Piernas & Abdomen:
1) Sentadilla con barra — 4x8-10
2) Prensa de piernas — 4x10-12
3) Peso muerto rumano — 4x8-10
4) Extensión de piernas — 3x15
5) Curl femoral — 3x15"""
            },
            "Arnold x UPL": {
                "Casa": """Rutina 4 – Arnold x UPL (Casa)
Día 1 — Upper:
1) Flexiones pies elevados — 4x10-12
2) Remo con mochila — 4x10-12
3) Fondos entre sillas — 3x12
4) Curl bíceps botellas — 3x12
5) Plancha con toques — 3x20 toques

Día 2 — Push:
1) Flexiones diamante — 4x12
2) Flexiones estándar — 3x15
3) Elevaciones frontales — 3x15
4) Elevaciones laterales — 3x15
5) Fondos en silla — 3x12

Día 3 — Legs:
1) Sentadillas con mochila — 4x15-20
2) Zancadas alternas — 4x12 c/pierna
3) RDL mochila — 3x12
4) Puente de glúteos — 3x20
5) Elevaciones talones — 3x20

Día 4 — Pull:
1) Remo toalla — 4x12
2) Superman — 3x15
3) Curl supino botellas — 3x12-15
4) Curl concentrado — 3x12
5) Plancha frontal — 3x45s

Día 5 — Core + Cardio:
1) Jumping Jacks — 4x45s
2) Crunch bicicleta — 3x20
3) Plancha lateral — 3x30s/side
4) Escaladores — 3x40s
5) V-ups — 3x15""",
                "Gimnasio": """Rutina 4 – Arnold x UPL (Gimnasio)
Día 1 — Upper:
1) Press de banca — 4x6-8
2) Dominadas lastradas — 4x8-10
3) Press militar — 3x8-10
4) Remo con barra — 3x10
5) Curl barra — 3x10-12
6) Extensión tríceps — 3x12

Día 2 — Push:
1) Press inclinado mancuernas — 4x10-12
2) Press máquina — 3x12
3) DB Military Press — 3x10-12
4) Elevaciones laterales — 3x15
5) Press francés — 3x12

Día 3 — Legs:
1) Sentadillas / Squats — 4x15-20
2) Zancadas — 4x12 c/pierna
3) RDL — 3x12
4) Puente de glúteos — 3x20
5) Calf Raises — 3x20

Día 4 — Pull:
(versión gym similar al pull casa con equipamiento)""",
            },
            "Bro Split": {
                "Casa": """Rutina 5 – Bro Split (Casa)
Día 1 — Pecho:
1) Flexiones clásicas — 4x15-20
2) Flexiones declinadas — 3x12-15
3) Flexiones diamante — 3x12
4) Aperturas con bandas — 3x15

Día 2 — Espalda:
1) Remo mochila — 4x15
2) Superman hold — 3x30s
3) Remo unilateral — 4x12-15

Día 3 — Piernas:
1) Sentadillas — 4x20
2) Zancadas — 3x12/pierna
3) Puente de glúteos — 4x20

Día 4 — Hombros:
1) Pike push-ups — 3x10-12
2) Elevaciones laterales — 3x15

Día 5 — Brazos:
1) Curl con botellas — 4x15
2) Fondos entre sillas — 4x12-15

Día 6 — Abdomen:
1) Crunch — 3x20
2) Elevaciones de piernas — 3x15""",
                "Gimnasio": """Rutina 5 – Bro Split (Gimnasio)
Día 1 — Pecho:
1) Press banca — 4x8-10
2) Press inclinado — 4x10-12
3) Aperturas — 3x12-15

Día 2 — Espalda:
1) Dominadas — 4x8-10
2) Remo con barra — 4x10
3) Pulldown — 3x12

Día 3 — Piernas:
1) Sentadilla — 4x8-10
2) Prensa — 4x10-12
3) Curl femoral — 3x15"""
            },
            "Torso/Pierna": {
                "Casa": """Rutina 6 – Torso/Pierna (Casa)
Día 1 — Torso A:
1) Flexiones — 4x15
2) Pike push-ups — 3x10-12
3) Fondos — 3x12
4) Plancha toques — 3x20

Día 2 — Piernas A:
1) Sentadilla mochila — 4x12
2) Zancadas — 3x12/pierna
3) RDL — 3x12
4) Calf Raises — 3x25""",
                "Gimnasio": """Rutina 6 – Torso/Pierna (Gimnasio)
Día 1 — Torso:
1) Press banca — 4x8-10
2) Dominadas — 4x8-10
3) Press militar — 3x8-10
4) Curl bíceps — 3x10-12

Día 2 — Pierna:
1) Sentadilla frontal — 4x8-10
2) Hip Thrust — 4x10-12
3) Curl femoral — 3x15
4) Calf Raise — 4x20"""
            },
            "Push Pull Legs": {
                "Casa": """Rutina 7 – Push Pull Legs (Casa)
Día 1 — Push:
1) Flexiones — 4x12-15
2) Elevaciones laterales — 3x15
3) Bench Dips — 3x12

Día 2 — Pull:
1) Remo mochila — 4x12
2) Curl bíceps — 3x12
3) Face pulls (banda) — 3x15

Día 3 — Legs:
1) Sentadillas — 4x15-20
2) RDL — 4x10
3) Calf Raises — 4x20""",
                "Gimnasio": """Rutina 7 – Push Pull Legs (Gimnasio)
Día 1 — Push:
1) Press banca — 4x8-10
2) Press inclinado — 4x10-12
3) Press militar — 4x10
4) Dips — 3x12

Día 2 — Pull:
1) Dominadas — 4x8-10
2) Remo con barra — 4x10
3) Curl bíceps — 4x10-12
4) Face pulls — 3x15

Día 3 — Legs:
1) Sentadilla frontal — 4x8-10
2) Hip Thrust — 4x10-12
3) Peso muerto convencional — 3x6-8"""
            }
        }

    def mostrar_rutina(self, nombre, variante):
        """Imprime la rutina formateada para 'nombre' y 'variante'."""
        rut = self.rutinas.get(nombre, {})
        texto = rut.get(variante)
        if not texto:
            print("No hay versión disponible para esta rutina/variante.")
            return
        print("\n" + "="*6 + f" {nombre} — {variante} " + "="*6)
        print(texto)
        print("="*30)

    def obtener_nombres_rutinas(self):
        return list(self.rutinas.keys())


class InterfazUsuario:
    def __init__(self):
        pass

    def pedir_entero(self, prompt, valid_range):
        while True:
            v = input(prompt + " ").strip().lower()
            if v == "volver":
                return "volver"
            try:
                v = int(v)
                if v in valid_range:
                    return v
                else:
                    print(" Opción inválida. Intenta de nuevo.")
            except ValueError:
                print("Introduce un número válido o escribe 'volver' para regresar.")

    def pedir_numero(self, prompt, tipo=float, positivo=True, limite_min=None, limite_max=None):
        while True:
            v = input(prompt + " ").strip().lower()
            if v == "volver":
                return "volver"
            try:
                v = tipo(v)
                if positivo and v <= 0:
                    print("Debe ser mayor que cero.")
                    continue
                if limite_min is not None and v < limite_min:
                    print(f"El valor mínimo permitido es {limite_min}.")
                    continue
                if limite_max is not None and v > limite_max:
                    print(f"El valor máximo permitido es {limite_max}.")
                    continue
                return v
            except ValueError:
                print("Valor inválido. Intenta otra vez o escribe 'volver' para regresar.")

    def pedir_si_no(self, prompt):
        print(prompt)
        print("1) Sí")
        print("2) No")
        r = self.pedir_entero("Elige 1 o 2:", {1, 2})
        if r == "volver":
            return "volver"
        return r == 1

    def convertir_peso_a_kg(self, peso, unidad):
        return peso if unidad == 1 else peso * 0.45359237

    def convertir_altura_a_m(self, altura, unidad):
        return altura / 100.0 if unidad == 1 else altura * 0.3048


class GestorRutinas:
    def __init__(self, rutina, interfaz):
        self.rutina = rutina
        self.interfaz = interfaz

    def menu_rutina_recomendada(self, sugerencia):
        """Menu interactivo para la rutina recomendada y navegación a otras rutinas."""
        while True:
            print("\n=== Rutina recomendada según tu plan ===")
            print(f"Rutina sugerida: {sugerencia}")
            print("1) Ver versión Casa")
            print("2) Ver versión Gimnasio")
            print("3) Otros ejercicios")
            print("4) Volver")
            opt = self.interfaz.pedir_entero("Elige 1-4:", {1,2,3,4})
            if opt == "volver":
                return  # vuelve al lugar desde donde fue llamado
            if opt == 1:
                self.rutina.mostrar_rutina(sugerencia, "Casa")
                sub = self.menu_post_vista_rutina()
                if sub == "volver":
                    continue
            elif opt == 2:
                self.rutina.mostrar_rutina(sugerencia, "Gimnasio")
                sub = self.menu_post_vista_rutina()
                if sub == "volver":
                    continue
            elif opt == 3:
                self.explorar_otras_rutinas()
            else:
                return

    def menu_post_vista_rutina(self):
        """Opciones después de ver una rutina: volver u otros ejercicios."""
        while True:
            print("\nOpciones:")
            print("1) Volver")
            print("2) Otros ejercicios")
            opt = self.interfaz.pedir_entero("Elige 1-2:", {1,2})
            if opt == "volver":
                return "volver"
            if opt == 1:
                return "volver"
            else:
                self.explorar_otras_rutinas()
                return "volver"

    def explorar_otras_rutinas(self):
        """Muestra la lista de rutinas disponibles y permite abrir cualquiera."""
        nombres = self.rutina.obtener_nombres_rutinas()
        while True:
            print("\n=== OTRAS RUTINAS DISPONIBLES ===")
            for i, n in enumerate(nombres, start=1):
                print(f"{i}) {n}")
            print(f"{len(nombres)+1}) Volver")
            opciones = set(range(1, len(nombres)+2))
            opt = self.interfaz.pedir_entero(f"Elige 1-{len(nombres)+1}:", opciones)
            if opt == "volver":
                return
            if opt == len(nombres)+1:
                return
            elegido = nombres[opt-1]
            # Elegir variante si hay
            print(f"\nHas elegido: {elegido}")
            variantes = list(self.rutina.rutinas[elegido].keys())
            for j, var in enumerate(variantes, start=1):
                print(f"{j}) {var}")
            print(f"{len(variantes)+1}) Volver")
            opciones_var = set(range(1, len(variantes)+2))
            var_choice = self.interfaz.pedir_entero(f"Elige 1-{len(variantes)+1}:", opciones_var)
            if var_choice == "volver":
                continue
            if var_choice == len(variantes)+1:
                continue
            variante = variantes[var_choice-1]
            self.rutina.mostrar_rutina(elegido, variante)
            input("\nPresiona ENTER para continuar en 'Otros ejercicios'...")


class PlanificadorEntrenamiento:
    def __init__(self, usuario, interfaz, gestor_rutinas):
        self.usuario = usuario
        self.interfaz = interfaz
        self.gestor_rutinas = gestor_rutinas

    def iniciar_plan(self):
        pasos = [
            "nombre", "sexo_edad", "peso", "altura", "actividad",
            "dias_rutina", "objetivo", "suplementos", "recordatorios", "resumen"
        ]
        paso_idx = 0

        while paso_idx < len(pasos):
            paso = pasos[paso_idx]

            # === NOMBRE ===
            if paso == "nombre":
                print("\n=== Paso 1: Identificación ===")
                nombre = input("Por favor, dime tu nombre (o escribe 'volver' para regresar): ").strip()
                if nombre.lower() == "volver":
                    paso_idx = max(0, paso_idx - 1)
                    continue
                self.usuario.nombre = nombre or "Usuario"
                print(f"Hola {self.usuario.nombre} , vamos a crear tu plan personalizado FitBoost.")
                paso_idx += 1

            # === SEXO Y EDAD ===
            elif paso == "sexo_edad":
                print("\n=== Paso 2: Datos personales ===")
                print("¿Eres?")
                print("1) Hombre")
                print("2) Mujer")
                sexo = self.interfaz.pedir_entero("Elige 1 o 2:", {1, 2})
                if sexo == "volver":
                    paso_idx -= 1
                    continue
                self.usuario.sexo = sexo
                edad = self.interfaz.pedir_entero("Ingresa tu edad (años, entre 10 y 100):", set(range(10, 101)))
                if edad == "volver":
                    paso_idx -= 1
                    continue
                self.usuario.edad = edad
                paso_idx += 1

            # === PESO ===
            elif paso == "peso":
                print("\n=== Paso 3: Peso ===")
                print("Elige unidad de peso:")
                print("1) kg")
                print("2) lb")
                unidad_peso = self.interfaz.pedir_entero("Elige 1 o 2:", {1, 2})
                if unidad_peso == "volver":
                    paso_idx -= 1
                    continue
                peso_raw = self.interfaz.pedir_numero("Ingresa tu peso:", float, limite_min=25, limite_max=250)
                if peso_raw == "volver":
                    paso_idx -= 1
                    continue
                self.usuario.peso_kg = self.interfaz.convertir_peso_a_kg(peso_raw, unidad_peso)
                print(f"Peso registrado: {self.usuario.peso_kg:.1f} kg")
                paso_idx += 1

            # === ALTURA ===
            elif paso == "altura":
                print("\n=== Paso 4: Altura ===")
                print("Elige unidad de altura:")
                print("1) cm")
                print("2) ft (pies, puedes usar decimales)")
                unidad_alt = self.interfaz.pedir_entero("Elige 1 o 2:", {1, 2})
                if unidad_alt == "volver":
                    paso_idx -= 1
                    continue
                altura_raw = self.interfaz.pedir_numero(
                    "Ingresa tu altura:",
                    float,
                    limite_min=100 if unidad_alt == 1 else 3.0,
                    limite_max=230 if unidad_alt == 1 else 7.5
                )
                if altura_raw == "volver":
                    paso_idx -= 1
                    continue
                self.usuario.altura_m = self.interfaz.convertir_altura_a_m(altura_raw, unidad_alt)
                print(f"Altura registrada: {self.usuario.altura_m*100:.1f} cm")
                paso_idx += 1

            # === ACTIVIDAD ===
            elif paso == "actividad":
                print("\n=== Paso 5: Nivel de actividad ===")
                niveles = {
                    1: ("Sedentario (poco o nada de ejercicio)", 1.2),
                    2: ("Ligero (1-3 días/sem)", 1.375),
                    3: ("Moderado (3-5 días/sem)", 1.55),
                    4: ("Activo (6-7 días/sem)", 1.725),
                    5: ("Muy activo (entrenamiento intenso diario)", 1.9)
                }
                for k, (desc, _) in niveles.items():
                    print(f"{k}) {desc}")
                opt = self.interfaz.pedir_entero("Elige 1-5:", set(niveles.keys()))
                if opt == "volver":
                    paso_idx -= 1
                    continue
                self.usuario.nivel_actividad, self.usuario.multiplicador_actividad = niveles[opt]
                paso_idx += 1

            # === RUTINA ===
            elif paso == "dias_rutina":
                print("\n=== Paso 6: Entrenamiento ===")
                nombre = self.usuario.nombre
                dias = self.interfaz.pedir_entero(f"{nombre}, ¿cuántos días a la semana puedes entrenar? (0-7):", set(range(0, 8)))
                if dias == "volver":
                    paso_idx -= 1
                    continue
                self.usuario.dias_entrenamiento = dias
                print(f"""
Rutinas disponibles:
-  Full Body → Ideal si entrenas 2-3 días/semana.
-  Upper/Lower → 3-4 días/semana.
-  Arnold Split → 5-6 días/semana.
-  Arnold x UPL → Variante avanzada con fuerza e intensidad.
-  Bro Split → 5-6 días, cada día un grupo muscular distinto.
-  Torso/Pierna → Equilibrio fuerza/hipertrofia.
-  Push Pull Legs → Estructura clásica PPL.
""")
                if dias <= 2:
                    sugerencia = "Full Body"
                elif dias <= 4:
                    sugerencia = "Upper/Lower"
                elif dias == 5:
                    sugerencia = "Arnold Split"
                else:
                    sugerencia = "Bro Split"
                self.usuario.rutina_sugerida = sugerencia
                print(f" Según tus días ({dias}), te sugerimos: {sugerencia}.")
                dur = self.interfaz.pedir_numero("¿Cuántos minutos entrenarás por sesión?:", int, limite_min=20, limite_max=180)
                if dur == "volver":
                    paso_idx -= 1
                    continue
                self.usuario.duracion_entrenamiento = dur
                paso_idx += 1

            # === OBJETIVO ===
            elif paso == "objetivo":
                print("\n=== Paso 7: Objetivo ===")
                objetivos = {
                    1: "Bajar de peso",
                    2: "Definir (reducir grasa, mantener músculo)",
                    3: "Ganar masa muscular"
                }
                for k, v in objetivos.items():
                    print(f"{k}) {v}")
                opt = self.interfaz.pedir_entero("Elige 1-3:", {1, 2, 3})
                if opt == "volver":
                    paso_idx -= 1
                    continue
                self.usuario.objetivo = objetivos[opt]
                paso_idx += 1

            # === SUPLEMENTOS ===
            elif paso == "suplementos":
                print("\n=== Paso 8: Suplementos ===")
                suplementos = self.interfaz.pedir_si_no("¿Deseas incluir suplementos en tu plan?")
                if suplementos == "volver":
                    paso_idx -= 1
                    continue
                if suplementos:
                    print("\nOpciones populares:")
                    print("Creatina: Optimum Nutrition, Creapure, MuscleTech")
                    print(" Proteína: Gold Standard, Dymatize ISO100, MyProtein")
                    print(" Pre-entreno: C4, Total War, Gorilla Mode")
                    presupuesto = self.interfaz.pedir_numero("¿Presupuesto mensual (en pesos colombianos COP)?", float, limite_min=20000, limite_max=2000000)
                    if presupuesto == "volver":
                        paso_idx -= 1
                        continue
                    self.usuario.usar_suplementos = True
                    self.usuario.presupuesto_suplementos = presupuesto
                else:
                    self.usuario.usar_suplementos = False
                    self.usuario.presupuesto_suplementos = 0
                paso_idx += 1

            # === RECORDATORIOS ===
            elif paso == "recordatorios":
                print("\n=== Paso 9: Recordatorios ===")
                agua = self.interfaz.pedir_si_no("¿Quieres recordatorios para beber agua?")
                if agua == "volver":
                    paso_idx -= 1
                    continue
                self.usuario.recordatorio_agua = agua
                if agua:
                    horas = self.interfaz.pedir_numero("¿Cada cuántas horas?", float, limite_min=0.5, limite_max=12)
                    if horas == "volver":
                        paso_idx -= 1
                        continue
                    self.usuario.intervalo_agua = horas
                else:
                    self.usuario.intervalo_agua = None

                ejercicio = self.interfaz.pedir_si_no("¿Recordatorios para ejercicio?")
                if ejercicio == "volver":
                    paso_idx -= 1
                    continue
                self.usuario.recordatorio_ejercicio = ejercicio
                if ejercicio:
                    horas_ej = self.interfaz.pedir_numero("¿Cada cuántas horas?", float, limite_min=1, limite_max=48)
                    if horas_ej == "volver":
                        paso_idx -= 1
                        continue
                    self.usuario.intervalo_ejercicio = horas_ej
                else:
                    self.usuario.intervalo_ejercicio = None
                paso_idx += 1

            # === RESUMEN FINAL ===
            elif paso == "resumen":
                altura_cm = self.usuario.altura_m * 100
                sexo, edad = self.usuario.sexo, self.usuario.edad
                peso, mult = self.usuario.peso_kg, self.usuario.multiplicador_actividad
                imc = self.usuario.calcular_imc()
                clasif = self.usuario.clasificar_imc()
                tdee, meta = self.usuario.calcular_calorias()

                print("\n=== 🧾 Resumen de tu plan FitBoost Colombia 🇨🇴 ===")
                print(f"Nombre: {self.usuario.nombre} | Edad: {edad} | Sexo: {'Hombre' if sexo == 1 else 'Mujer'}")
                print(f"IMC: {imc:.2f} ({clasif})")
                print(f"Rutina sugerida: {self.usuario.rutina_sugerida} ({self.usuario.dias_entrenamiento} días/sem, {self.usuario.duracion_entrenamiento} min/sesión)")
                print(f"Objetivo: {self.usuario.objetivo}")
                print(f"TDEE: {tdee:.0f} kcal | Meta diaria: {meta:.0f} kcal")
                print(f"Suplementos: {'Sí' if self.usuario.usar_suplementos else 'No'} | Presupuesto: ${self.usuario.presupuesto_suplementos:.0f} COP")
                print(f"Recordatorios: Agua cada {self.usuario.intervalo_agua}h | Ejercicio cada {self.usuario.intervalo_ejercicio}h")
                print("\nPlan completado. ¡Sigue constante y disciplinado! ")

                # --- Aquí ofrecemos ver la rutina recomendada ---
                while True:
                    print("\nOpciones finales:")
                    print("1) Ver rutina recomendada")
                    print("2) Volver al menú principal")
                    opt_final = self.interfaz.pedir_entero("Elige 1 o 2:", {1,2})
                    if opt_final == "volver":
                        # si el usuario escribió 'volver' volvemos al paso anterior (resumen no tiene más atrás)
                        continue
                    if opt_final == 1:
                        self.gestor_rutinas.menu_rutina_recomendada(self.usuario.rutina_sugerida)
                        # una vez el usuario termine de explorar rutinas, seguimos al menú principal
                        break
                    else:
                        break

                break

        return self.usuario


class FitBoostApp:
    def __init__(self):
        self.interfaz = InterfazUsuario()
        self.rutina = Rutina()
        self.gestor_rutinas = GestorRutinas(self.rutina, self.interfaz)
        self.usuario_actual = None

    def main(self):
        while True:
            print("\n========== MENÚ PRINCIPAL ==========")
            print("1) Iniciar nuevo plan FitBoost")
            print("2) Ver resumen del último plan")
            print("3) Salir")
            opcion = self.interfaz.pedir_entero("Elige una opción:", {1, 2, 3})

            if opcion == 1:
                usuario = Usuario()
                planificador = PlanificadorEntrenamiento(usuario, self.interfaz, self.gestor_rutinas)
                self.usuario_actual = planificador.iniciar_plan()
            elif opcion == 2:
                if self.usuario_actual:
                    print("\n=== Último resumen guardado ===")
                    self.mostrar_resumen_usuario()
                    # dar opción para ver la rutina recomendada desde el resumen guardado
                    while True:
                        print("\nOpciones:")
                        print("1) Ver rutina recomendada del último plan")
                        print("2) Volver")
                        o = self.interfaz.pedir_entero("Elige 1 o 2:", {1,2})
                        if o == "volver":
                            break
                        if o == 1:
                            self.gestor_rutinas.menu_rutina_recomendada(self.usuario_actual.rutina_sugerida)
                            break
                        else:
                            break
                else:
                    print("\nAún no tienes un plan guardado. Inicia uno nuevo primero.")
                input("\nPresiona ENTER para continuar.")
            else:
                print("\nGracias por usar FitBoost Colombia. ¡Hasta pronto! 🏋️‍♂️")
                break

    def mostrar_resumen_usuario(self):
        if not self.usuario_actual:
            return

        usuario = self.usuario_actual
        imc = usuario.calcular_imc()
        clasif = usuario.clasificar_imc()
        tdee, meta = usuario.calcular_calorias()

        print(f"Nombre: {usuario.nombre}")
        print(f"Edad: {usuario.edad}")
        print(f"Sexo: {'Hombre' if usuario.sexo == 1 else 'Mujer'}")
        print(f"Peso: {usuario.peso_kg:.1f} kg")
        print(f"Altura: {usuario.altura_m*100:.1f} cm")
        print(f"IMC: {imc:.2f} ({clasif})")
        print(f"Nivel de actividad: {usuario.nivel_actividad}")
        print(f"Días de entrenamiento: {usuario.dias_entrenamiento}")
        print(f"Duración por sesión: {usuario.duracion_entrenamiento} minutos")
        print(f"Rutina sugerida: {usuario.rutina_sugerida}")
        print(f"Objetivo: {usuario.objetivo}")
        print(f"TDEE: {tdee:.0f} kcal | Meta diaria: {meta:.0f} kcal")
        print(f"Suplementos: {'Sí' if usuario.usar_suplementos else 'No'}")
        if usuario.usar_suplementos:
            print(f"Presupuesto suplementos: ${usuario.presupuesto_suplementos:.0f} COP")
        print(f"Recordatorio agua: {'Cada ' + str(usuario.intervalo_agua) + ' horas' if usuario.recordatorio_agua else 'No'}")
        print(f"Recordatorio ejercicio: {'Cada ' + str(usuario.intervalo_ejercicio) + ' horas' if usuario.recordatorio_ejercicio else 'No'}")


if __name__ == "__main__":
    app = FitBoostApp()
    app.main()
```
